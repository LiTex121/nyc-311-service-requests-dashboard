```python
import pandas as pd

df = pd.read_csv("../data/raw/311_Service_Requests_from_2020_to_Present_20260611.csv")
```

```python
# picking important columns
print(df.columns.to_list())

df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
print(df.columns.to_list())

Columns_To_Keep = ['unique_key', 'created_date', 'closed_date', 'agency', 'agency_name', 'borough', 'problem_(formerly_complaint_type)', 'status']

df_clean = df[Columns_To_Keep].copy()
```

```python
# Renaming a few

df_clean = df_clean.rename(columns={
    'unique_key' : 'request_id',
    'problem_(formerly_complaint_type)' : 'complaint_type'

    })

df_clean
```

```python
# Remove Dups and nulls and check for spaces, etc.

df_clean = df_clean.dropna(subset=[
    "request_id",
    "created_date",
    "agency",
    "agency_name",
    "borough",
    "complaint_type",
    'status'
]) # skiped over closed_date because not all case's are closed.
df_clean = df_clean.drop_duplicates()
df_clean.shape
```

```python
# cleaning date format for better calulations later on.

df_clean["created_date"] = pd.to_datetime(df_clean["created_date"], errors="coerce")
df_clean["closed_date"] = pd.to_datetime(df_clean["closed_date"], errors="coerce")

df_clean
```
