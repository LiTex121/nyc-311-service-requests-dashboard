```python
import pandas as pd

df = pd.read_csv("../data/raw/311_Service_Requests_from_2020_to_Present_20260611.csv")
```

```python
# Renaming a few

df_clean = df_clean.rename(columns={
    'unique_key' : 'request_id',
    'problem_(formerly_complaint_type)' : 'complaint_type'

    })

df_clean
```
